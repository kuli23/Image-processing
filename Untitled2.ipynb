```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('natasha_grayscale.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None
t1 = np.linspace(0, 100, 101).astype('uint8')  
t2 = np.linspace(125, 249, 125).astype('uint8')
t3 = np.linspace(225, 255,30).astype('uint8') 
transform1 = np.concatenate((t1,t2,t3), axis=0).astype('uint8')
fig, axis = plt.subplots()
fig.suptitle("Curve")
axis.plot(transform1)
axis.set_xlabel(r'X axis, Input')
axis.set_ylabel(r'Y axis, Output')
axis.set_xlim(0,255)
axis.set_ylim(0,255)
axis.set_aspect('equal')
axis.grid()
plt.show()
image_transform1 = cv.LUT(img, transform1)
fig, axis = plt.subplots(1,2, figsize=(15,25))
axis[0].imshow(img, cmap="gray")
axis[0].set_title("Original Image")
axis[1].imshow(image_transform1, cmap="gray")
axis[1].set_title("Intensity Transformed Image")
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('spider.png', cv.IMREAD_COLOR)
assert img is not None
im23 = cv.cvtColor(img, cv.COLOR_BGR2HSV)
h,s,v = cv.split(im23)
fig, axis1= plt.subplots(1,3, figsize=(15,25))
axis1[0].imshow(h, cmap="gray")
axis1[0].set_title('Hue_image')
axis1[1].imshow(s, cmap="gray")

axis1[1].set_title('Saturation_image')
axis1[2].imshow(v, cmap="gray")
axis1[2].set_title('Value_image')
plt.show()
x= np.arange(0, 256).astype('uint8')
a = .3
sigma = 70
y = np.minimum(((x)+(a*(np.exp(-(x-128)**2/(2*sigma**2))))/128), 255).astype('uint8')
image_transform2 = cv.LUT(s, y)
plt.title('Transformation Curve of Intensity')
plt.plot(y)
plt.show()
newHSV = cv.merge([h,image_transform2,v])
result = cv.cvtColor(newHSV, cv.COLOR_HSV2BGR)
added_img = cv.add(newHSV, img)
fig, axis= plt.subplots(1,3, figsize=(15,25))
axis[0].imshow(im23, cmap="gray")
axis[0].set_title('Original Image')
axis[1].imshow(newHSV, cmap="gray")
axis[1].set_title('Transformation of vibrance')
axis[2].imshow(added_img, cmap="gray")
axis[2].set_title('Vibrance Enhanced Image')
plt.show()
```

```python
#gamma used= 0.7
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread ('highlights_and_shadows.jpg', cv.IMREAD_COLOR)
assert img is not None
img_LAB = cv.cvtColor(img, cv.COLOR_BGR2LAB)
L_img, a_img , b_img = cv.split(img_LAB)
gamma = .7
t = np.array([(i/255.)**gamma*255 for i in range (256)], np.uint8)
g = t[L_img]
plt.suptitle("Gamma correction")
plt.plot(t)
plt.show()
fig, axis = plt.subplots(1,2, figsize=(8,3))
fig.suptitle("b. Applying gamma correction")
axis[0].imshow(L_img, cmap="gray")
axis[0].set_title("Original Image")
axis[1].imshow(g,cmap="gray")
axis[1].set_title("Gamma Corrected Image")
plt.show()
plt.figure(figsize = [8, 4])
plt.subplot(1, 2, 1)
plt.gca().set_title('Original')
img_h = cv.calcHist([L_img],[0],None,[256],[0,256])
plt.plot(img_h)
plt.subplot(1, 2, 2)
plt.gca().set_title('Gamma Corrected')
g_h = cv.calcHist([g],[0],None,[256],[0,256])
plt.plot(g_h)
plt.show()
```

```python

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('washed_out_aerial_image.png', cv.IMREAD_GRAYSCALE)
assert img is not None
plt.figure(figsize = [8, 4])
plt.subplot(1, 2, 1)
plt.gca().set_title('Original')
h = np.zeros(256)
h = [np.sum(img==i) for i in range (256)]
plt.bar(range(256), h)
plt.subplot(1, 2, 2)
plt.gca().set_title('After Equalization')
eh = cv.equalizeHist(img)
plt.hist(eh)
plt.show()
fig, ax= plt.subplots(1,2, figsize=(10,20))
ax[0].imshow(img, cmap="gray")
ax[0].set_title('Original Figure')
ax[1].imshow(eh, cmap="gray")
ax[1].set_title('After Equalization Figure')
plt.show()
```

```python
thresh = cv.inRange(saturated_img, 15, 230)
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (20,20))
morph = cv.morphologyEx(thresh, cv.MORPH_CLOSE, kernel)
mask = morph
result = cv.bitwise_and(img, img, mask=mask)

fig, ax = plt.subplots(1,3, figsize=(10,2.5))
fig.suptitle("b) Extracting Foreground mask")
ax[0].imshow(img, cmap="gray")
ax[0].set_title("Original")
ax[1].imshow(mask, cmap="gray")
ax[1].set_title("Foreground Mask")
ax[2].imshow(result, cmap="gray")
ax[2].set_title("Foreground Image")
plt.show()
```
